```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('Data/Melbourne01.csv', on_bad_lines = "skip", low_memory=False)
df
```

```python
# New list of column names
new_names = ['year', 'month', 'day', 'hour', 'minute', 'Air Temp (degrees C)', 'Apparent Temp (degrees C)', 'Dew Pt Temp (degrees C)',
            'Humidity (%)', 'Wind Direction', 'Wind Speed (km/h)', 'Wind Gust (km/h)', 'MSLP (hPa)', 'Rainfall since 9 am (mm)']

# Assign the new names
df.columns = new_names
df.head()
```

```python
df.describe()
```

## Deal with Shifted Columns

combine fixed shifted_column dataframe with the normal (but not complete) dataframe

```python
shifted_df = pd.read_csv('Data/Melbourne01-fixxing.csv', low_memory = False)
shifted_df
```

```python
shifted_df.columns = new_names
shifted_df.head()
```

```python
all_df = pd.concat([df,shifted_df], join='outer')
all_df = all_df.sort_index()
all_df
```

## Deal with Duplicated Rows

```python
# all_df['day'] = pd.to_numeric(all_df['day'], errors='coerce')
# all_df['hour'] = pd.to_numeric(all_df['hour'], errors='coerce')
# all_df['minute'] = pd.to_numeric(all_df['minute'], errors='coerce')
all_df['Air Temp (degrees C)'] = pd.to_numeric(all_df['Air Temp (degrees C)'], errors='coerce')
all_df['Apparent Temp (degrees C)'] = pd.to_numeric(all_df['Apparent Temp (degrees C)'], errors='coerce')
all_df['Dew Pt Temp (degrees C)'] = pd.to_numeric(all_df['Dew Pt Temp (degrees C)'], errors='coerce')
all_df['Humidity (%)'] = pd.to_numeric(all_df['Humidity (%)'], errors='coerce')
# all_df['Wind Direction'] = all_df['Wind Direction'].astype(str)
# df['Wind Direction'] = df['Wind Direction'].str.strip()
all_df['Wind Speed (km/h)'] = pd.to_numeric(all_df['Wind Speed (km/h)'], errors='coerce')
all_df['Wind Gust (km/h)'] = pd.to_numeric(all_df['Wind Gust (km/h)'], errors='coerce')
all_df['MSLP (hPa)'] = pd.to_numeric(all_df['MSLP (hPa)'], errors='coerce')
all_df['Rainfall since 9 am (mm)'] = pd.to_numeric(all_df['Rainfall since 9 am (mm)'], errors='coerce')
all_df['Rainfall since 9 am (mm)'] = all_df['Rainfall since 9 am (mm)'].fillna(0)
all_df
```

```python
print(all_df.duplicated().sum())
```

```python
all_df_fixed = all_df.drop_duplicates()
all_df_fixed = all_df_fixed.reset_index(drop=True)
all_df_fixed
```

## Deal with clustered rows

```python
print(all_df_fixed.isna().sum())
```

```python
need_fix_rows = np.where(all_df_fixed['day'].isna())[0].tolist()
# print(need_fix_rows)
print(len(need_fix_rows))
```

```python
good_rows = all_df_fixed.index.difference(need_fix_rows)
len(good_rows)
```

### Good Rows DF

```python
good_df = all_df_fixed.loc[good_rows]
good_df
```

```python
good_df['Timestamp'] = pd.to_datetime(good_df[['year','month','day','hour','minute']])
good_df.head()
```

```python
good_df.drop(columns = ['year','month','day','hour','minute'], inplace=True)
good_df.head()
```

```python
good_df.index = good_df['Timestamp']
good_df.drop(columns = ["Timestamp"], inplace=True)
good_df.head()
```

### Bad Rows DF

```python
bad_df = all_df_fixed.loc[need_fix_rows]
bad_df
```

```python
len(bad_df) # Number of all the bad rows
```

```python
no_month = np.where(all_df_fixed['month'].isna())[0].tolist() # Select rows with no data in month column
with_month = [n for n in need_fix_rows if n not in no_month] # Select rows with data clustered in month column
```

```python
bad_w_month = all_df_fixed.loc[with_month] # Create a dataframe for bad rows with data in month column
len(bad_w_month)
```

```python
bad_no_month = all_df_fixed.loc[no_month] # Create a dataframe for bad rows with NO data in month column
len(bad_no_month)
```

32+1232 = 1264 which is the exact number of row with missing values

#### Dealing with bad_df (without month)

```python
bad_fixed = bad_no_month['year'].astype(str).str.split('\t', expand=True)
bad_fixed.drop(columns=[14], inplace=True)
bad_fixed.head()
```

```python
bad_fixed.columns = [
    'year', 'month', 'day', 'hour', 'minute',
    'Air Temp (degrees C)', 'Apparent Temp (degrees C)', 'Dew Pt Temp (degrees C)',
    'Humidity (%)', 'Wind Direction', 'Wind Speed (km/h)',
    'Wind Gust (km/h)', 'MSLP (hPa)', 'Rainfall since 9 am (mm)'
]
```

```python
bad_fixed.head()
```

```python
bad_fixed['Timestamp'] = pd.to_datetime(bad_fixed[['year','month','day','hour','minute']])
bad_fixed.drop(columns = ['year','month','day','hour','minute'], inplace=True)
bad_fixed.index = bad_fixed['Timestamp']
bad_fixed.drop(columns = ['Timestamp'], inplace=True)
bad_fixed.head()
```

#### Dealing with bad_w_month (bad with month)

```python
split_bad_w_month = bad_w_month['month'].astype(str).str.split('\t', expand=True)
split_bad_w_month.head()
```

```python
year_for_bad_w_month = bad_w_month['year']
year_for_bad_w_month = pd.DataFrame(year_for_bad_w_month)
year_for_bad_w_month.head()
```

```python
bad_w_month_fixed = pd.merge(year_for_bad_w_month, split_bad_w_month, left_index=True, right_index=True, how='right')
bad_w_month_fixed.columns = new_names
bad_w_month_fixed.head()
```

```python
bad_w_month_fixed['Timestamp'] = pd.to_datetime(bad_w_month_fixed[['year','month','day','hour','minute']])
bad_w_month_fixed.drop(columns = ['year','month','day','hour','minute'], inplace=True)
bad_w_month_fixed.index = bad_w_month_fixed['Timestamp']
bad_w_month_fixed.drop(columns = ['Timestamp'], inplace=True)
bad_w_month_fixed.head()
```

### Merge them back together

```python
df_fixed = pd.concat([good_df, bad_fixed, bad_w_month_fixed])
df_fixed = df_fixed.sort_index()
df_fixed
```

```python
print(df_fixed.isna().values.any())
```

#### Right here, all the data are in the same format (timestamp, Air Temp, Apparent Temp, Dew Pt Temp, ....) but this doesn't mean that the data is all clean.
There might be some duplicate, missing time period, error values (outliers), etc.
We still ned to check for those and try to fix them.

## Deal with Duplicated Rows (after cleaning)

```python
df_fixed['Air Temp (degrees C)'] = pd.to_numeric(df_fixed['Air Temp (degrees C)'], errors='coerce')
df_fixed['Apparent Temp (degrees C)'] = pd.to_numeric(df_fixed['Apparent Temp (degrees C)'], errors='coerce')
df_fixed['Dew Pt Temp (degrees C)'] = pd.to_numeric(df_fixed['Dew Pt Temp (degrees C)'], errors='coerce')
df_fixed['Humidity (%)'] = pd.to_numeric(df_fixed['Humidity (%)'], errors='coerce')
df_fixed['Wind Direction'] = df_fixed['Wind Direction'].astype(str)
df_fixed['Wind Direction'] = df_fixed['Wind Direction'].str.strip()
df_fixed['Wind Speed (km/h)'] = pd.to_numeric(df_fixed['Wind Speed (km/h)'], errors='coerce')
df_fixed['Wind Gust (km/h)'] = pd.to_numeric(df_fixed['Wind Gust (km/h)'], errors='coerce')
df_fixed['MSLP (hPa)'] = pd.to_numeric(df_fixed['MSLP (hPa)'], errors='coerce')
df_fixed['Rainfall since 9 am (mm)'] = pd.to_numeric(df_fixed['Rainfall since 9 am (mm)'], errors='coerce')
df_fixed.tail()
```

```python
print(df_fixed.duplicated().sum())
```

```python
df_fixed = df_fixed.drop_duplicates()
df_fixed.tail()
```

```python
print(df_fixed.duplicated().sum())
```
