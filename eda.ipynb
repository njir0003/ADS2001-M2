```python
df_melb = pd.read_csv("data/Melbourne01.csv", on_bad_lines='skip')
df_adl = pd.read_csv("data/Adelaide01.csv", on_bad_lines='skip')
df_dew_point = pd.read_excel("data/dew point.xlsx")
```

```python
cols = ['year', 'month', 'day', 'hour', 'minute', 'temp', 'apparent_temp',
        'dew_point', 'humidity', 'wind_dir', 'wind_speed', 'wind_gust',
        'pressure', 'rainfall_since_9am']
```

```python
# Apply column names defined above
df_melb.columns = cols

# Coerce date/time columns to numeric (drops corrupted rows)
for col in ['year','month','day','hour','minute']:
    df_melb[col] = pd.to_numeric(df_melb[col], errors='coerce')

df_melb.dropna(subset=['year','month','day','hour','minute'], inplace=True)

# Build datetime and set as index (required for resample to work)
df_melb['datetime'] = pd.to_datetime(
    df_melb[['year','month','day','hour','minute']].astype(int)
)
df_melb = df_melb.drop(columns=['year','month','day','hour','minute'])
df_melb = df_melb.set_index('datetime').sort_index()

print(f'Shape: {df_melb.shape}')
print(f'Date range: {df_melb.index.min()} → {df_melb.index.max()}')
df_melb.head()
```

```python
# Encode wind direction as sin/cos before resampling
compass_to_deg = {
    'N':0,'NNE':22.5,'NE':45,'ENE':67.5,
    'E':90,'ESE':112.5,'SE':135,'SSE':157.5,
    'S':180,'SSW':202.5,'SW':225,'WSW':247.5,
    'W':270,'WNW':292.5,'NW':315,'NNW':337.5,
    'CALM':np.nan,'VAR':np.nan
}

deg = df_melb['wind_dir'].astype(str).str.strip().str.upper().map(compass_to_deg)
df_melb['wind_dir_sin'] = np.sin(np.deg2rad(deg))
df_melb['wind_dir_cos'] = np.cos(np.deg2rad(deg))

# Drop the original string column — can't be resampled
df_melb = df_melb.drop(columns=['wind_dir'])

print('Wind direction encoded.')
df_melb[['wind_dir_sin','wind_dir_cos']].dropna().head()
```

```python
met_cols = ['temp','apparent_temp','dew_point','humidity',
            'wind_speed','wind_gust','pressure','rainfall_since_9am']

df_melb[met_cols] = df_melb[met_cols].replace(-9999, np.nan)


n_at_zeros = (df_melb['apparent_temp'] == 0).sum()
df_melb['apparent_temp'] = df_melb['apparent_temp'].replace(0, np.nan)
print(f'apparent_temp zeros replaced: {n_at_zeros}')


n_p_zeros = (df_melb['pressure'] == 0).sum()
df_melb['pressure'] = df_melb['pressure'].replace(0, np.nan)
print(f'pressure zeros replaced: {n_p_zeros}')


n_rain_null = df_melb['rainfall_since_9am'].isnull().sum()
df_melb['rainfall_since_9am'] = df_melb['rainfall_since_9am'].fillna(0)
print(f'rainfall NaNs filled with 0: {n_rain_null}')

print(f'\nRemaining NaNs (filled by interpolation in next step):')
print(df_melb[met_cols].isnull().sum().to_string())
```

```python
# Resample to uniform 30-min grid, then interpolate gaps
df_melb = (
    df_melb.select_dtypes(include='number')
    .resample('30min')
    .mean()
    .interpolate(method='time')
)
print(f'Shape after resampling: {df_melb.shape}')
df_melb.head()
```
